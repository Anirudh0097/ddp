## Verifying data collection and procedure followed by *scipy.integrate*
* ode
* odeint

```python
global y_odeint, y_ode
```

```python
from scipy.integrate import ode, odeint
import matplotlib.pyplot as plt
```

```python
def derv_ode(t,y):
    y_ode.append([float(t),float(y)])
    return t**2

def derv_odeint(y,t):
    y_odeint.append([float(t),float(y)])
    return t**2
```

```python
y0=.5
y_odeint = []
time  = numpy.linspace(0, 2.5, 10)

soln_odeint = odeint(derv_odeint, y0, time)
```

```python
y0=.5
t0 = 0.0
t1 = 1.0
y_ode = []
solver = ode(derv_ode)
solver.set_initial_value( y0 , t0 )
sol = []
while solver.successful() and solver.t < t1:
    solver.integrate( t1 , step = True )
    sol.append([solver.t,solver.y])
```

```python
time_odeint = time
ans_odeint = soln_odeint

time_odeint_all = numpy.array( [ float( y_odeint[i][0] ) for i in range( len( y_odeint )) ] )
ans_odeint_all = numpy.array( [float( y_odeint[i][1] ) for i in range( len( y_odeint ) ) ] )

time_ode = numpy.array( [ float( sol[i][0] ) for i in range( len(sol) ) ] )
ans_ode = numpy.array( [ float( sol[i][1] ) for i in range( len( sol ) )] )

time_ode_all = numpy.array( [ float( y_ode[i][0] ) for i in range( len( y_ode )) ] )
ans_ode_all = numpy.array( [ float( y_ode[i][1] ) for i in range( len( y_ode )) ] )
```

```python
plot_odeint, = plt.plot( time_odeint, ans_odeint,'m--'  )
#plot_odeint_all, = plt.plot( time_odeint_all, ans_odeint_all, 'b|' )
plot_ode, = plt.plot( time_ode, ans_ode, 'g.-')
plot_ode_all, = plt.plot( time_ode_all, ans_ode_all, 'r+')
plt.legend( (plot_odeint, plot_odeint_all, plot_ode, plot_ode_all) , ('odeint', 'odeint-all', 'ode', 'ode-all') , 'lower right' )
xlabel(r'$time$')
plt.grid()
plt.figsize(10,10)
plt.show()
```

* odeint-all prepares the path needed for odeint's engine to cover
